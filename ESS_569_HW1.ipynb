```python
# Importing required modules
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# Declare a basic Python list and a NumPy array
my_list = [1, 2, 3, 4, 5]
my_array = np.array([1, 2, 3, 4, 5])

# Print list and array
print("List:", my_list)
print("NumPy Array:", my_array)

# Declare a basic Python function
def square_elements(elements):
    """Returns the square of each element in a list or NumPy array."""
    return [element**2 for element in elements]

# Call the function with both the list and NumPy array
squared_list = square_elements(my_list)
squared_array = square_elements(my_array)

# Print the results
print("Squared List:", squared_list)
print("Squared NumPy Array:", squared_array)

# Create a simple Matplotlib figure and save it as a PNG
x = np.linspace(0, 10, 100)  # Generate 100 points between 0 and 10
y = np.sin(x)  # Compute the sine of each point

plt.figure(figsize=(8, 6))  # Create a figure with size 8x6 inches
plt.plot(x, y, label="Sine Wave")  # Plot the sine wave
plt.title("Sine Wave Example")  # Add a title
plt.xlabel("x values")  # Label the x-axis
plt.ylabel("sin(x)")  # Label the y-axis
plt.legend()  # Add a legend
plt.grid(True)  # Add a grid for better visibility

# Save the figure as a PNG file
plt.savefig("sine_wave_example.png")

# Show the plot in the notebook
plt.show()
```
